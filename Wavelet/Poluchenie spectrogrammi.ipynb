```python
# Библеотеки для работы с вейвлетами, массивами и графиками
import pywt
import numpy as np
import matplotlib.pyplot as plt

# Тестовый сигнал с двумя составляющими
fs = 1000  # Частота дискретизации
t = np.arange(0, 1, 1/fs) 
f1 = 5
f2 = 50
x = np.sin(2*np.pi*f1*t) 
#+ np.sin(2*np.pi*f2*t)
# Применяем вейвлет-преобразование для получения спектрограммы с использованием вейвлета Морле
cwtmatr, freqs = pywt.cwt(x, np.arange(1, 140), 'morl')
# Yflj dct gthtltkfnm
# Отображаем спектрограмму
plt.figure(figsize=(20, 10))
plt.imshow(cwtmatr, extent=[0, 1, 1, 140], cmap='PRGn', aspect='auto',vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
plt.colorbar(label='Амплитуда')
plt.title('Спектрограмма сигнала')
plt.xlabel('Время (с)')
plt.ylabel('Частота (Гц)')
plt.show()
```

```python
# Это уже с вейвлетом-мексиканской шляпой
cwtmatr, freqs = pywt.cwt(x, np.arange(1, 100), 'mexh')

# Отображаем спектрограмму
plt.figure(figsize=(20, 5))
plt.imshow(cwtmatr, extent=[0, 1, 1, 100], cmap='PRGn', aspect='auto',vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
plt.colorbar(label='Амплитуда')
plt.title('Спектрограмма сигнала')
plt.xlabel('Время (с)')
plt.ylabel('Частота (Гц)')
plt.show()
```
