```python
#     ,   
import pywt
import numpy as np
import matplotlib.pyplot as plt

#     
fs = 1000  #  
t = np.arange(0, 1, 1/fs) 
f1 = 5
f2 = 50
x = np.sin(2*np.pi*f1*t) 
#+ np.sin(2*np.pi*f2*t)
#  -       
cwtmatr, freqs = pywt.cwt(x, np.arange(1, 140), 'morl')
# Yflj dct gthtltkfnm
#  
plt.figure(figsize=(20, 10))
plt.imshow(cwtmatr, extent=[0, 1, 1, 140], cmap='PRGn', aspect='auto',vmax=abs(cwtmatr).max(), vmin=-abs(cwtmatr).max())
plt.colorbar(label='')
plt.title(' ')
plt.xlabel(' ()')
plt.ylabel(' ()')
plt.show()
```
